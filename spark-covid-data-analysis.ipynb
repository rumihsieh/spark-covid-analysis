```scala
// Monthly Positivity Rate (Statewide)
val positivity_by_month = data_rdd_new
  .map{ case (date, _, cases, tests) =>
    val ym = date.slice(0,7)        // take only "YYYY-MM"
    (ym, (cases, tests))
  }

  .reduceByKey{ (a, b) =>
    (a._1 + b._1, a._2 + b._2) //for the same month
  }

  .map{ case (ym, (cases, tests)) =>
    val positivity = (cases.toDouble / tests) * 100
    (ym, Math.round(positivity * 100).toDouble / 100)
  }
  .sortBy(_._2, ascending = false)
  .collect()
```

```scala
//Daily Positivity per County (Safe Division & Option)

// It handles division by zero safely
def divide(numerator: Int, denominator: Int): Option[Double] = {
  if (denominator == 0) None
  else Some(numerator.toDouble / denominator)
}

```

```scala

val positivity = data_rdd_new.map{
  case (date, county, cases, tests) =>

    val ratio = divide(cases, tests)

    // convert Option[Double] to a number (0.0 if None)
    val positivity_value = ratio match {
      case Some(r) => Math.round(r * 10000).toDouble / 100 //keep 2 decimal if Math.round(r * 100).toDouble would be 13.0
      case None    => 0.0
    }

    // return (date, county, positivity)
    (date, county, positivity_value)
}
.collect()

```

```scala
//Highest Positive Day in Dataset
val highest_positive = positivity_1.takeOrdered(1)(Ordering[Double].reverse.on(_._3))
```

```scala
//Monthly Positivity Rate by County
val county_month_rdd = data_rdd_new.map{
  case (date, county, cases, tests) =>
    val month = date.slice(0,7)      // get "YYYY-MM"
    ((county, month), (cases, tests))
}

val summed_rdd = county_month_rdd.reduceByKey{
  case ((cases1, tests1), (cases2, tests2)) =>
    (cases1 + cases2, tests1 + tests2)
}

val positivity_rdd = summed_rdd.map{
  case ((county, month), (cases, tests)) =>
    val positivity = if (tests == 0) 0.0 else (cases.toDouble / tests)
    (county, month, Math.round(positivity * 100).toDouble / 100) 
}

val sorted_rdd = positivity_rdd.sortBy(r => (r._1, r._2))

sorted_rdd.collect().foreach{
  case (county, month, positivity) =>
    println(s"$county $month $positivity")
}
```
